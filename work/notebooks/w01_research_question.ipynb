## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

The starter dataset contains 30,000 pages. About 54.2% of the pages are labelled as declining, showing that decline is common enough to justify a prioritization system. The dataset also contains pages with very different levels of visibility, so a ranked review queue could help editors focus first on declining pages with greater practical importance.

```python
import os
import sys
import subprocess
import pandas as pd

repo_dir = "flyrank-ml-internship-starter"

if not os.path.isdir(repo_dir):
    subprocess.run(
        [
            "git",
            "clone",
            "--depth",
            "1",
            "https://github.com/flyrank-bih/flyrank-ml-internship-starter",
            repo_dir
        ],
        check=True
    )

os.chdir(repo_dir)

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

total_pages = len(df)
declining_rate = (
    df["trend_direction"]
    .str.lower()
    .eq("down")
    .mean()
)
median_impressions = df["impressions_90d"].median()

print("Total pages:", total_pages)
print("Declining rate:", round(declining_rate, 3))
print("Median impressions in 90 days:", round(median_impressions, 1))
```
